```python
from tqdm import tqdm

lambd_range = [0.5,0.6,0.7,0.8,0.9]
starting_state = ((0,0),(1,2))
V_plot=[]

for lambd in tqdm(lambd_range):
    V, policy,V_t = mz.value_iteration(env, lambd, epsilon)
    V_plot.append(V_t)
plt.figure(figsize=(15,10))
for i in range(len(lambd_range)):
    plt.plot(V_plot[i],label='$\lambda$ = '+str(lambd_range[i]));
plt.xlabel('No of iterations')
plt.ylabel('Value function')
plt.legend()
plt.show()
```
