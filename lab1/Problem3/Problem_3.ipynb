### SARSA algorithm using $\epsilon$-greedy exploration

```python
from tqdm import tqdm
lambd = 0.8
maxIter = 10000000
eps_range = [0.1,0.15,0.2,0.25,0.5]
starting_state = ((0,0),(3,3))
V_plot=[]
for eps in tqdm(eps_range):
    Q, V_t = bk.sarsa(bank, lambd, eps,starting_state, maxIter)
    V_plot.append(V_t)
plt.figure(figsize=(15,10))
for i in range(len(eps_range)):
    plt.plot(V_plot[i],label='$\epsilon$ = '+str(eps_range[i]));
plt.xlabel('No of iterations')
plt.ylabel('Value function')
plt.legend()
plt.show()
```
